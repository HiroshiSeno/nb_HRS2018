```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import os
import subprocess
%matplotlib inline
```

# APDを計測

```python
vmem_310 = np.zeros(300)
vmem_308 = np.zeros(300)
vmem_306 = np.zeros(300)
vmem_304 = np.zeros(300)
vmem_302 = np.zeros(300)
```

```python
for i in range(300):
    num = "{0:05d}".format(i+8701)
    vmem_310[i] = np.load('/mnt/recordings/SimulationResults/2D/20171020-1/%s/v.npy'%num)
    vmem_308[i] = np.load('/mnt/recordings/SimulationResults/2D/20171020-2/%s/v.npy'%num)
    vmem_306[i] = np.load('/mnt/recordings/SimulationResults/2D/20171020-3/%s/v.npy'%num)
    vmem_304[i] = np.load('/mnt/recordings/SimulationResults/2D/20171020-4/%s/v.npy'%num)
    vmem_302[i] = np.load('/mnt/recordings/SimulationResults/2D/20171020-5/%s/v.npy'%num)
```

## 白黒

```python
plt.plot(vmem_310, label='310K', color='black')
plt.plot(vmem_308, label='308K', color='black', linestyle='--')
plt.plot(vmem_306, label='306K', color='black', linestyle='-.')
plt.plot(vmem_304, label='304K', color='black', linestyle=':')
# plt.plot(vmem_302, label='302K', color='black', linestyle=':')
plt.legend(fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Time (ms)', fontsize=16)
plt.ylabel('Membrane potential (mV)', fontsize=16)
plt.title('Prolongation of action potential duration', fontsize=18)
plt.tight_layout()
# plt.savefig('/mnt/recordings/AnalysisResult/Seno/simulation/Mahajan/20171125/APD_cooling.png', dpi=300)
# plt.close()
```
